```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
df = pd.read_csv('regression_datasets_large.xlsx - Electricity_Bill (2).csv')
df.head()
```

```python
x = df[['Units_Consumed']]
y = df['Bill_Amount']

model = LinearRegression()
model.fit(x,y)
```

```python
print(model.coef_[0])
m = model.coef_[0]
```

```python
print(model.intercept_)
c = model.intercept_
```

```python
y_pred = model.predict(x)
print(y_pred)
```

```python
import matplotlib.pyplot as plt


plt.scatter(x, y, color='blue', label='Actual Data')


plt.plot(x, y_pred, color='red', label='Regression Line')

units = 350
bill_pred = m * units + c
plt.scatter([units], [bill_pred], color='red', s=100, label='Prediction Point')


plt.text(units, bill_pred, f'  Predicted Bill = {bill_pred:.2f}', color='black', fontsize=10)


plt.xlabel("Units Consumed")
plt.ylabel("Bill Amount")
plt.title("Electricity Bill Prediction")
plt.legend()
plt.show()
```
